```python
import os

%matplotlib inline
import matplotlib.image as mpimg
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
base_dir = "ImaÌgenes-data-science-ii"

print("Contents of base directory:")
print(os.listdir(base_dir))

print("\nContents of train directory:")
print(os.listdir(f'{base_dir}/train'))

print("\nContents of test directory:")
print(os.listdir(f'{base_dir}/test'))
```

```python
train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")
```

```python
train_baseball_dir = os.path.join(train_dir, "baseball")
train_cricket_dir = os.path.join(train_dir, "cricket")
train_football_dir = os.path.join(train_dir, "football")
```

```python
train_baseball_fnames = os.listdir(train_baseball_dir)
train_cricket_fnames = os.listdir(train_cricket_dir)
train_football_fnames = os.listdir(train_football_dir)

print("Baseball picture names -> ",  train_baseball_fnames[:10], "\n")
print("Cricket picture names-> ", train_cricket_fnames[:10], "\n")
print("Football picture names -> ", train_football_fnames[:10], "\n")
```

```python
# Total number of each category in the training directory 

print("Total training baseball images:", len(os.listdir(train_baseball_dir) ))
print("Total training cricket images:", len(os.listdir(train_cricket_dir) ))
print("Total training football images:", len(os.listdir(train_football_dir) ))
```

```python
# Set up matplotlib fig, and size it to fit 4x4 pics
nrows = 4
ncols = 4

pic_index = 0
fig = plt.gcf()
fig.set_size_inches(ncols*4, nrows*4)

pic_index+=8

next_baseball_pix = [os.path.join(train_baseball_dir, fname) 
                for fname in train_baseball_fnames[pic_index-8:pic_index] 
               ]

next_cricket_pix = [os.path.join(train_cricket_dir, fname) 
                for fname in train_cricket_fnames[pic_index-8:pic_index]
               ]

next_football_pix = [os.path.join(train_football_dir, fname) 
                for fname in train_football_fnames[pic_index-8:pic_index]
               ]

for i, img_path in enumerate(next_baseball_pix + next_cricket_pix + next_football_pix):
    # Set up subplot; subplot indices start at 1
    sp = plt.subplot(nrows, ncols, i+1)
    sp.axis('Off') # Don't show axes (or gridlines)
    img = mpimg.imread(img_path)
    plt.imshow(img)

plt.show()
```

```python
##### The model
```

```python
model = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 150x150 with 3 bytes color
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2), 
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'), 
    tf.keras.layers.MaxPooling2D(2,2),
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(), 
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'), 
    # Only 1 output neuron. It will contain a value from 0-1 where 0 for 1 class ('cats') and 1 for the other ('dogs')
    tf.keras.layers.Dense(3, activation="softmax")  
])
```

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics = ["accuracy"])
```

```python
train_datagen = ImageDataGenerator(rescale = 1.0/255.)
test_datagen  = ImageDataGenerator(rescale = 1.0/255.)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    batch_size=20,
                                                    class_mode="sparse",
                                                    target_size=(150, 150)) 
test_generator = test_datagen.flow_from_directory(test_dir)
```

```python
history = model.fit(
            train_generator,
            epochs=15,
            validation_data=test_dir,
            verbose=2
            )
```
